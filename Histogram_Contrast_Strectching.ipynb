# Calculate histogram Using Library

calculate histogram using library and display.

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#calculating histogram

image1 = cv2.imread('../../standard_test_images/cameraman.tif',0)
image2 = cv2.imread('../../standard_test_images/peppers_color.tif',1)
```

```python
plt.imshow(image1[:,:],cmap='gray')
```

```python
#converting BGR-->RGB
#matplotlib takes RGB input
plt.figure()
plt.imshow(image2[:,:,::-1])
plt.title('Colored Image')
#converting all images to gray
grayImage2 = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)

plt.figure()
plt.imshow(grayImage2, cmap = 'gray')
plt.title('Greyed Image')
```

```python
#calcluate using cv2.calcHist

#cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
hist1 = cv2.calcHist([image1], [0], None, [256], [0,256])

hist2 = cv2.calcHist([grayImage2], [0], None, [256], [0,256])

fig, axs = plt.subplots(2,1, figsize = (13,8))
axs[0].plot(hist1)
axs[0].set_title('Image 1 Histogram')
axs[1].plot(hist2)
axs[1].set_title('Image 2 Histogram')

```

# Calcluate using Own Implementation

```python
#hist3 for image1
#hist4 for greyed image2

hist3 = np.zeros((256,1))
hist4 = np.zeros((256,1))

#iterating over image1
for intensityVal in image1.ravel():
    hist3[intensityVal,0] += 1

#iterating over image1
for intensityVal in grayImage2.ravel():
    hist4[intensityVal,0] += 1
```

```python
plt.figure(figsize = (9,4))
plt.plot(hist3)
plt.title('Image1 Histogram(Own Implementation)')

plt.figure(figsize = (9,4))
plt.plot(hist4)
plt.title('Image2 Histogram(Own Implementation)')
```

# Contrast Strectching using Library


```python
#reading images

darkImage1 = cv2.imread("Images/image4.jpg", 0)
darkImage2 = cv2.imread("Images/image6.jpg", 0)


```

```python

plt.figure()
fig, axs = plt.subplots(1,2, figsize = (20, 20))

axs[0].set_title('Dark Image1')
axs[0].imshow(darkImage1, cmap = 'gray')

axs[1].set_title('Dark Image2')
axs[1].imshow(darkImage2, cmap = 'gray')

```

```python
#using library 
stretchedDarkImage1 = cv2.equalizeHist(darkImage1)
stretchedDarkImage2 = cv2.equalizeHist(darkImage2)


plt.figure(figsize = (30,30))

plt.subplot(221)
plt.hist(darkImage1.flatten(),256,[0,256], color = 'r')

plt.subplot(222)
plt.hist(darkImage2.flatten(),256,[0,256], color = 'r')

plt.subplot(223)
plt.hist(stretchedDarkImage1.flatten(),256,[0,256], color = 'b')

plt.subplot(224)
plt.hist(stretchedDarkImage2.flatten(),256,[0,256], color = 'b')

plt.show()
```
